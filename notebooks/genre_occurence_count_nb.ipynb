## model 

```python
print("new model")

model = make_pipeline(
    TfidfVectorizer(
        max_df = 0.9, 
        min_df = 3, 
        ngram_range = (1, 3),   
        norm = 'l1',
        stop_words = 'english'
    ), 
    MultinomialNB(alpha = 1e-2, fit_prior = False)
)


model.fit(training_data, training_target)

predicted_genres = model.predict(test_data)
```

```python
# multi-class confusion matrix
mat = confusion_matrix(test_target, predicted_genres)
sns.heatmap(mat.T, square = True, annot = True, fmt = "d", xticklabels = keep_genres, yticklabels = keep_genres)

plt.xlabel("true lables")
plt.ylabel("predicted label")

fig = plt.gcf()
fig.set_size_inches(20, 11)

print(f"the accuracy is {accuracy_score(test_target, predicted_genres)}")
```
